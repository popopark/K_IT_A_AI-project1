# **2. 문자열 안에서 특수한 값을 나타내기**

제어 문자를 활용하여 특수한 값을 나타낼 수 있다.

* \n : 엔터 값을 나타낸다.
* \n : Tab 값을 나타낸다.
* \' : 작은 따옴표를 나타낸다.
* \" : 큰따옴표를 나타낸다.
* \\\ : \\를 나타낸다.

```python
print('오늘은 금\n요일')
print('껌 \t 200원')
print('과자 \t 500원')  # \t은 탭 키를 누른 것 처럼 일정 간격을 띄어준다.

print('나는 생각했다 \'배고프다\'라고')

print("나는 말했다 \"배고프다\" 라고")

print('happy \\new \year')
```

# **3. 문자열의 형식을 지정하는 방법**



## **3-1. %연산자를 활용**

* int --> %d (형식 지정자)
* float --> %f (형식 지정자)
* str --> %s (형식 지정자)

      옵션
        %[0 or -][n][.m]형식지정자
        ex) %06.2f
        0 : 비어있는 공간을 0으로 채워줌(생략시 띄어쓰기)
        - : 왼쪽 정렬 (기본은 오른쪽 정렬)

        n : 전체 자릿수
        .m : 소수점 아래 자릿수

```python
print('오늘은 %02d월 %10d일 입니다.' % (2, 15))
print('원주율은 %6.1f 입니다.' % 3.14)
print('나의 이름은 %10s 입니다.' % '박준')

print('오늘은 %2d월 %-d일 입니다.' % (2, 15))
print('원주율은 %.1f 입니다.' % 3.14)
print('나의 이름은 %s 입니다.' % '박준')
```

# **5. 문자열 관련 함수와 메소드**

* len(문자열) : 함수 사용 결과는 문자열의 길이


메소드
* count() : 함수 사용 결과는 문자열 안에서 특정 문자의 갯수  
* find() : 함수 사용 결과는 문자열 안에서 특정 문자의 위치(인덱스 번호)  
* upper() : 함수 사용 결과는 대문자로 바뀐 문자열
* strip() : 함수 사용 결과는 앞뒤 띄어쓰기가 없어진 문자열
* replace() : 함수 사용 결과는 특정 문자를 특정 문자로 바꾼 문자열
* split() : 함수 사용 결과는 문자열이 분리된 리스트
* join() : 함수 사용 결과는 리스트 요소를 연결시킨 하나의 문자열
