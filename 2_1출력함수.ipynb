# **2. 값의 종류**

값의 종류는 **type** 이라고 부르기도 하며, **자료형** 이라고 부르기도 한다.

*   **int**

> 10, 20, -50, ....(정수)

*   **float**

> 3.1, -2.4, ....(실수)

*   **string**

> '안녕하세요', '반갑습니다', 'apple', ....(문자열)

*   **bool**
*   **list**
*   **tuple**
*   **dict**
*   **set**
