##**1-2. 리스트 함수와 메소드**



```python
# len(반복가능객체), 함수 사용 결과는 반복 가능 객체 속 요소의 갯수 (int가 결과)

len([10, 15, '안녕하세요'])

li4 = []
li4

# append(요소) : 리스트 마지막에 요소를 추가한다, 함수 사용 결과가 값이 아니다(None)

li4.append('박준') # append 함수는 추가 해주라는 명령의 함수다.
print(li4)

li4.append('박준1')
li4.append('박준1')
li4.append('박준1')
print(li4.append('박준2'))
print(li4)

# insert(인덱스, 요소)
li4.insert(0, '안녕') # 0번째 방 안에다가 '안녕'을 추가 해줘 라는 내용
print(li4)

# remove(요소), 리스트에서 요소를 삭제해준다 (가장 앞에 있는 요소)
li4.remove('안녕') # 없는 요소를 삭제 할 때는 오류가 난다.
print(li4)

li4.pop(0) # pop은 index 번호로 삭제한다.
print(li4)

# pop(인덱스), 리스트에서 인덱스 위치의 요소를 삭제(기본은 마지막)
#               함수 사용 결과가 삭제된 요소 값이다.

print(li4.pop())
print(li4)

# del 변수를 삭제 할 때 사용

name = '박준'
del name # del을 쓴 다음에 뒤에 변수를 쓰면 삭제를 시켜 준다.

# print(name)

name2 = '홍길동'
name = name2

print(li4)
del li4[2] # li4[2]방에 존재하는 값을 메모리에서 해제(삭제)
print(li4)
del li4[0:1]
print(li4)
```

```python
li5 = [10, 20, 31, 10, 20, 10]
li5.count(10) 

# count는 요소로, 함수 사용 결과는 리스트 속에서 요소의 갯수(int)

# help(함수 이름) --> 괄호 안에있는 함수의 설명을 보여준다.
# index(요소, 시작 인덱스, 끝 인덱스) 리스트 안에서 요소의 위치 인덱스를 찾는다, 시작 인덱스 부터 끝 인덱스 -1 사이에서
# 시작 인덱스와 끝 인덱스 생략 시 처음부터 끝가지 함수 사용 결과는 찾은 요소의 위치 인덱스(int)

li5.index(20) # index는 앞에 있는 숫자 부터 찾게 된다.
print(li5.index(20, 2, 5))
help(li5.index)

print(li5)
li5.reverse() # 역순으로 정렬해주는 함수
print(li5)


# sort(key = 정렬 방식, reverse = True/False)
# 기본은 오름차순으로 정렬, reverse True : 내림차순 정렬, key에 함수를 넘겨주면 함수가 적용된 상태로 정렬

li5.sort() # 오름차순으로 정렬해주는 함수
print(li5)


li6 = ['apple', 'pineapple', 'grape', 'banana']
li6.sort(key = len)

print(li6)

help(li6.sort)
```

```python
li7 = [3, 9, 2]

sorted(li7) # 함수 사용 결과가 정렬된 리스트, 원본은 바꾸지 않는다.

print(li7)

li7.sort()
print(li7)
```

```python
numbers = [23, 56, 15, 20]
# extend(반복 가능 객체), 원래 리스트에다가 괄호안에 있는 반복 가능 객체를 합쳐 준다.

numbers.extend([3, 5, 6, 7]) # number + extend 개념념
print(numbers)

numbers.extend('안녕하세요')
print(numbers)
```

# **2. 반복 가능 객체의 형 변환**

* list(반복 가능 객체)
 > 함수 사용 결과가 괄호 안에 있는 반복 가능 객체가 리스트로 변환된 결과
 
* tuple(반복 가능 객체)
 > 함수 사용 결과가 괄호 안에 있는 반복 가능 객체가 tuple로 변환된 결과


```python
list('안녕하세요')

tuple([10, 20, 20, 30])
```

##**2-1. 여러개의 int 값을 띄어 쓰기로 입력 받기**

```python
# 숫자 두 개를 입력 받아서 두 수의 총합 구하기

nums = input('숫자 두 개를 띄어쓰기로 입력 >>').split()

# nums = ['10', '30']

print(f'결과 : {int(nums[0]) + int(nums[1])}')
print(nums)
```

```python
nums = list(map(int, input('숫자 두개 >> ').split())) # map 타입은 반복 가능 객체체

print(nums)
```

```python
# map()함수
# map(함수, 반복 가능 객체)
# map 함수 사용 결과는 반복 가능 객체의 요소에 각각 함수를 적용시킨 반복 가능 객체이다.

list(map(int, ['10', '20', '30']))

# ['10', '20', '30']
# li[0] = int(li[0])
# li[1] = int(li[1])
# li[2] = int(li[2])
# print(li)
```
