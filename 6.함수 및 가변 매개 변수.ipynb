```python
def changeNum(target):

  hangeul = '공일이삼사오육칠팔구'
  
  res = ''
  for ch in target:
    res += str(hangeul.index(ch))

  return res 
```

#**1. 함수의 return**

* 함수 사용 결과를 **값**으로 사용하고 싶다면 함수 내부에 return을 작성하면 된다.  
* return 옆에 값이 있는 함수는 함수 사용 결과를 return된 값으로 치환 하여 해석한다.  
* 함수 사용 결과가 값이 아닌 함수는 None 타입의 함수다.  

```python
def add(a, b):
  print(a + b)
  print(f'a에 들어 있는 값 {a}')
  print(f'b에 들어 있는 값 {b}')

  return a + b
```

```python
a = print(add(10, 20) / 2)  # add 함수가 사용되면 가장 안쪽 부터 사용됌.

print(a)                    # 함수 사용 결과가 값이 아닌 함수는 None 값이 return 된다.
```

#**2. 가변 매개 변수**

* 매개 변수의 갯수가 가변적으로 변하는 매개 변수

```python
def f1(a, b):
  print(f' a속 값 {a}')
  print(f' b속 값 {b}')
```

```python
f1(10, 20)
```

```python
def f2(*a):
  print('f2 실행 됌')
```

```python
# a라는 매개 변수에 여러개의 값이 대입되는 것이 아니라
# a라는 매개 변수에는 여러개의 값이 요소로 들어있는 튜플 하나가 대입되는 것이다.

def f2(*a):       # a는 변수임
                  # 가변 매개 변수를 튜플로 생각하고 활용하면 된다.

  print('f2 실행 됨')
  print(a)
  print(type(a))
  print('_____')
```

```python
f2(10)
f2(10, 20)
f2(10, 20, 30)
```

```python
# 가변 매개 변수가 맨 뒤에 있는 경우에는
# 인자로 넘어간 값들이 순서대로 일반 매개 변수에 대입되고 나머지 튜플로 묶여 가변 매개 변수에 대입된다.

# 가변 매개 볍ㄴ수가 가운데 있거나 앞에 있는 경우에는 일반 매개 변수에 값이 대입되지 않는다.
# (튜플로 묶여서 가변 매개 변수가) 가져가기 때문에, 따라서 이럴 경우에는 인자를 넘겨줄때 일반 매개 변수 이름을 직정 작성한다.

def f3(a, b, *c):   
  print(a)
  print(b)
  print(c)
  print('_____')
```

```python
f3(10, 20)
f3(10, 20, 30, 40)
```

#**6. return 값이 여러개인 함수**

```python
# 여러개의 값이 return 되는 것이 아니라
# 여러개의 값이 요소로 들어있는 하나의 tuple이 return 되는 것이며
# 함수 사용 결과는 tuple 타입 이라고 생각한다.

def f6():
  print('f6 함수가 실행 됌')
  return '박준', 10, 183.3        # 소괄호가 생략 됌
```

```python
name, age, height = f6()

print(name, age, height)
print('프로그램 종료')
```

#**9. 익명 함수(람다 함수)**

함수의 이름 없이 함수를 정의하는 방법, 일회성으로 사용 할 때 만드는 함수

```python
def f(a):
  return a * 20

lambda a : a * 20     # 이름 없는 함수임
```

```python
# 익명 함수를 만들면서 사용하기

(lambda a : a * 20)(10)
f(10)
```

```python
# 익명 함수의 매개 변수가 두 개인 경우

lf1 = lambda a, b : a+b
lf1(10, 20)

# 익명 함수의 매개 변수가 없는 경우

lf2 = lambda : print('안녕')
lf2()
```

```python
myList = ['banana', 'pineapple', 'grape', 'apple', 'peach']

print(myList)

# 일회성 함수
myList.sort(key = lambda a: a.count('a'))

print(myList)
```
