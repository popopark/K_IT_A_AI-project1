# **1. 리스트**

* 순서가 있다.
* 중복된 값을 저장할 수 있다.
* 수정 가능하다(mutable)

## **1-1. 리스트 생성하기**

1. list() 함수를 활용하기
2. [] 를 활용하기


```python
a = [10, 15, 35, '안녕', 3.14]
type(a)

a


# 요소 조회
# [] 인덱스 연산자를 활용한다.

a[0] # 0은 0번째 방 = 10 / 1은 1번째 방 = 15
a[4]
a[-1]


# 인덱스 슬라이싱
a[1:3]
```

```python
# 비어있는 (요소가 0개인) 리스트 만들기

li1 = []

type(li1) # list 타입

li1 # []

li2 = list()
type(li2) # list 타입

li2 # []
```

```python
li3 = [1, '안녕', ['홍길동', '김철수'], 3.14] # 구분을 만들어 주는 것이 중요함!

li3[2]
li3[2][1] = '김영희' # 김철수는 [ 방 0 , 방 1] 방1에 들어 있기 때문에 [2] 리스트 안에 [1] 리스트를 불러내면 김철수가 나온다.
                     # 김영희 이름을 넣는 것도 동일함.

li3
```

```python
scores = [80, 50, 60] # 국어, 영어, 수학

scores[1] = 100 # scores [1] 저장 공간에 100을 넣는다.

scores # scores를 불러 내면 저장 공간에 있는 최신 숫자가 출력 됌.
```
