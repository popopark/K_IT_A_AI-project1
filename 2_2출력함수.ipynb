# **4. print() 함수의 부가 기능**





## **4-1. 여러개의 값을 동시에 출력하기**

* print() 소괄호 안쪽에 출력하고 싶은 값들을, (콤마) 로 나열한다.

```python
# '박준', 10, 3.14 세 개의 값을 동시에 출력하기

print('박준', 10, 3.14)
```

##**4-2. print() 함수에서 여러개의 값을 출력할 때 옵션 조절하기**

* 소괄호 안쪽의 sep(매개변수, parameter)에 연결하고 싶은 값을 전달한다.


```python
print(10, '안녕', '오늘은', '금요일')
```

```python
print(10, '안녕', '오늘은', '금요일', sep = '') # sep(세퍼레이트) (파라미터)는 옵션 기능을 한다.
```

## **4-3. print() 함수 실행 후 엔터 값 조정하기**

* 소괄호 안쪽의 end (매개변수, parameter)에 엔터 대신에 사용하고 싶은 값을 전달한다.


```python
print(10, end = 'aaaa') # 출력한 뒤에 마지막에 출력할 값을 설정(기본값은 enter 값)
print(20, end = ' ')
print(30)
```

```python
# ent 옵션과 sep 옵션 함께 사용하기
# 순서는 중요하지 않다!

print(10, 20, 30 , 40, sep = '🎁🐱‍🚀😎', end = '반가워') # 이모티콘은 윈도우 키 + 점
print(100, 200, 300)

```
