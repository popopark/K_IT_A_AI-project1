# **1. 산술 연산자**

1. + : 덧셈 연산자
2. - : 뺄셈 연산자
3. * : 곱셈 연산자
4. / : 나눗셈 연산자

# **1-1. 덧셈 연산자**


```python
# 10 + 5 # 15, int + int --> int
# 10 + 3.14 # 13.14, int + float --> float
# 10 + '안녕' # TypeError : unsupported operand type(s) for + : 'int' and 'str'

# float + str, str + float --> 오류
# 3.5 + 1.5 # 5.0 , float + float --> float


print('안녕' + '반가워') # ;안녕 반가워', str + str --> str 연결된다.




```

```python
print(type(10))
print(type('안녕'))
```

## **1-2. 뺄셈 연산자**

# str은 뺄셈 불가능
# 피연산자 어느 한 쪽이라도 float 타입이 있다면 결과는 float 타입
# int - int -> int

```python
type(10 - 5) # 5, int - int --> int

type(10 - 0.0) # 10.0, int - float, float - int --> float

# str은 뺄셈 연산 불가능
# type(10 - '10')
type(1.2 - 0.2) # 1.0, float - float --> float
```
