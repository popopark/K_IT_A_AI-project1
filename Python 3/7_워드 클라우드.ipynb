### 문제
konlpy에서 제공되는 constitution.txt를 읽어 명사를 추출하고 빈도수에 따라 워드 클라우드를 작성해보자

```python
!pip install konlpy
```

```python
from konlpy.corpus import kolaw
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
text
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
noun_text = okt.nouns(text)
print(noun_text)
```

```python
noun_text.sort(key=lambda x: len(x))
print(noun_text)
```

```python
# stop_words = [~~~~~~~~]
# noun_text = [each_word for each_word in noun_text if each_word not in stop_words]
noun_text = [each_word for each_word in noun_text if len(each_word) > 1]
print(noun_text)
```

```python
from collections import Counter
```

```python
count = Counter(noun_text)
print(count)
```

```python
data = count.most_common(500)
data = dict(data)
print(data)
```

```python
# 딕셔너리에 저장된 데이터를 워드 클라우드로 생성하라면 generate_from frequencies() 메소드를 사용해야 함
wordcloud = WordCloud(
    max_words=500,
    background_color='ivory',
    font_path='/usr/share/fonts/truetype/nanum/NanumPen.ttf'
).generate_from_frequencies(data)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='Kaiser')
plt.axis('off')
plt.show()
```
