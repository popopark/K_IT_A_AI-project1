```python
import folium
```

```python
# folium.Map(location=[위도, 경도], zoom_start=배열, ...)

map_folium = folium.Map(location=[37.518965, 127.025855], zoom_start=17)
map_folium
```

```python
# Marker(): location 속성으로 지ㅓㅈㅇ한 위치에 마커를 생성
# popup(): 마커를 클릭 했을 때 표시 할 문자열을 설정
# add_to(): 작성된 마커를 지도에 추가

popup = folium.Popup('코리아IT아카데미', max_widty=200)
map_folium = folium.Map(location=[37.50007, 127.03559], zoom_start=17)
folium.Marker(location=[37.50007, 127.03559], popup=popup).add_to(map_folium)
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)


popup = folium.Popup('서울특별시청', max_widty=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup).add_to(map_folium)

popup = folium.Popup('덕수궁', max_widty=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_widty=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup).add_to(map_folium)

popup = folium.Popup('광화문', max_widty=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_widty=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup).add_to(map_folium)


map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)


popup = folium.Popup('서울특별시청', max_widty=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup,
              icon=folium.Icon(color='red', icon='star')).add_to(map_folium)

popup = folium.Popup('덕수궁', max_widty=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup,
              icon=folium.Icon(color='pink', icon='ok')).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_widty=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup,
              icon=folium.Icon(color='orange', icon='star')).add_to(map_folium)

popup = folium.Popup('광화문', max_widty=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup,
              icon=folium.Icon(color='blue', icon='user')).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_widty=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup,
              icon=folium.Icon(color='green', icon='star')).add_to(map_folium)


map_folium
```

```python
# CircleMarker(): 원의 범위를 마커로 생성

map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)


popup = folium.Popup('서울특별시청', max_widty=200)
folium.CircleMarker(location=[37.56652479270633, 126.9779129116262], popup=popup,
              radius=100, color='red', fill_color='#EC407A').add_to(map_folium)

map_folium
```
