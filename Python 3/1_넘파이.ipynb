#2. 행렬 연산

##2-1. 연산자

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
# 덧셈 연산
# shape가 같아야 함
# 같은 position 끼리 연산

a + b
```

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4],
              [1, 2]])
```

```python
# a + b ValueError: operands could not be broadcast together with shapes (2,3) (2,2) 
```

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])
b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
# 뺄셈 연산
a - b
```

```python
# 곱셈 연산
a + b
```

```python
# 나눗셈 연산
a / b
```

```python
# dot product(행렬의 내적)
# 맞닿는 shape가 같아야 함
# 떨어져 있는 shape가 결과 행렬이 됌

a = np.array([[1, 2, 3],
              [1, 2, 3],
              [2, 3, 4]])
b = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
a.shape, b.shape
```

```python
print((1*1+2*3+3*5), (1*2+2*4+3*6))
print((1*1+2*3+3*5), (1*2+2*4+3*6))
print((2*1+3*3+4*5), (2*2+3*4+4*6))
```

```python
np.dot(a, b)
```

```python
# 문자열 역순으로 출력하기

s = 'python'
print(s[::-1])
print(s[3:0:-1]) # 3번 인덱스부터 0직전 까지 역순으로 1개씩 가져오기
print(s[3::-1]) # 3번 인덱스부터 0까지 역순으로 1개씩 가져오기
print(s[::-1]) # 마지막부터 0까지 역순으로 1개씩 가져오기
```

```python
ndarr2d = np.array([[11, 10, 12, 9],
                    [3, 1, 4 ,2],
                    [5, 6, 7, 8]])
```

```python
ndarr2d.shape
```

```python
# 행 정렬

np.sort(ndarr2d, axis=0)
```

```python
# 열 정렬

np.sort(ndarr2d, axis=1)
```

```python
# 축의 마지막 방향

np.sort(ndarr2d, axis=-1)
```
