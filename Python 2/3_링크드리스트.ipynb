##**1. 링크드리스트(Linked List)**

* 떨어진 곳에 존재하는 데이터를 화살표로 연결해서 관리하는 구조
* 데이터의 삽입과 삭제가 매우 빠름
* C언어에서는 중요한 자료 구조지만, 파이썬에서는 리스트 타입이 링크드 리스트 역할도 모두 지원

###**1-1.링크드리스트의 용어**

* 노드(node) : 데이터 저장 단위 및 위치. 데이터와 포인터로 구성
* 포인터(pointer) : 각 노드 안에서 다음이나 이전의 노드와의 연결 정보를 가지고 있는 공간

```python
# 링크드리스트 구현
# 링크드리스트를 구현 할 때 클래스를 주로 활용

class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
```

```python
node1 = Node(10)
node2 = Node(20)

print(node1)
print(node2)

node1.next = node2
head = node1
```

```python
print(node1)
print(head)
```

##**2.객체 지향 프로그래밍으로 링크드리스트 구현하기**


```python
class Node:
  def __init__(self, data, next=None):
    self.data = data
    self.next = next
```

```python
class NodeMgmt:
  def __init__(self, data):
    self.head = Node(data)

  def add(self, data):
    if self.head == None:
        self.head = Node(data)
    else:
        node = self.head
        while node.next:
            node = node.next
        node.next = Node(data)

  def node_print(self):
      node = self.head
      while node:
          print(node.data)
          node = node.next

  def delete(self, data):
    if self.head.data == data:
      temp = self.head
      self.head = self.head.next
      del temp
    else: # 아니라고 한다면
      node = self.head
      while node.next:
        if node.next.data == data:
          temp = node.next
          node.next = node.next.next
          del temp
          return      # 함수 빠져 나가는 것
        else:
          node = node.next
```

```python
linkedlist1 = NodeMgmt(0)
linkedlist1.node_print()
```

```python
for data in range(1, 11):
  linkedlist1.add(data)
```

```python
linkedlist1.node_print()
```

```python
linkedlist1.delete(5)
```
