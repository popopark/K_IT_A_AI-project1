# 1. 순차 탐색(Sequential Search)

* 탐색이란 여러 데이터 중에서 원하는 데이터를 찾아내는 것을 의미  

* 순차 탐색이란 데이터가 담겨 있는 리스트를 앞에서 부터 차례대로 하나씩 비교하여 원하는 데이터를 찾는 방법

```python
import random
```

```python
rand_data_list = list()

for num in range(10):
    rand_data_list.append(random.randint(1, 100))

rand_data_list
```

```python
def sequencial(data_list, search_data):
    for index in range(len(data_list)):
        if data_list[index] == search_data:
             return search_data
    return -1
```

```python
sequencial(rand_data_list, 100)
```

```python
sequencial(rand_data_list, 25)
```

```python
sequencial(rand_data_list, 90)
```
