# 1. 너비 우선 탐색(Breadth-First Search)

### 1-1. BFS란?

* 대표적인 그래프 탐색 알고리즘
* 정점들과 같은 레벨에 있는 노드(형제 노드)들을 먼저 탐색하는 방식
* 한 단계씩 내려오면ㅁ서 해당 노드와 같은 레벨에 있는 노드들을 먼저 순회함

```python
graph = dict()

# {'A':['B', 'C']}
graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']

graph
```

### 1-2. BFS 알고리즘 구현

* 큐 자료 구조를 활용
* need_visit 큐, visited 큐 생성

```python
data = [10, 20, 30, 40]
data.pop()
print(data)
```

```python
data.pop(0) # 큐
print(data)
```

```python
def bfs(graph, start_node):
  visited, need_visit = list(), list()

  need_visit.append(start_node)

  while need_visit:
    node = need_visit.pop(0)
    if node not in visited:
        visited.append(node)
        need_visit.extend(graph[node])
    
  return visited
```

```python
bfs(graph, 'A')
```
