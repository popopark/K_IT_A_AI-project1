#**2. 생성자**

* 클래스명()
* 해당 클래스 타입의 객체를 생성해주는 역할을 한다.
* 생성자를 직접 정의하여 인스턴스 변수를 인스턴스 마다 똑같게 만들어 줄 수가 있다.
* 생성자는 파이썬에서 '__ init __()' 로 관리한다.

```python
class Dog:

  # 생성자를 직접 정의, 생성자가 실행 되면 정의된 대로 동작한다.

  def __init__(self, name ='', age =0):     # __ 언더바가 붙은 함수는 파이썬에서 특별 하게 관리하는 함수임

    # 생성자를 직접 정의하여 인스턴스(객체)를 생성할 때 객체 속에 들어있는
    # 인스턴스 변수를 공통적으로 만들어 줄 수 있다.

    self.name = name
    self.age = age

  # print('생성자 실행!')

  def show(self):
    print(f'저의 이름은 {self.name}입니다')
    print(f'저의 나이는 {self.age}살 입니다')
    
```

```python
d1 = Dog('뽀삐', 5)
d1.show()

d2 = Dog('백구', 6)
d2.show()

d3 = Dog()

d3.name = input()
d3.age = int(input())
```

```python
class MyDate:
  def __init__(self, year, month, date, hour = 0, min = 0, sec = 0):
    self.year = year
    self.month = month
    self.date = date
    self.hour = hour
    self.min = min
    self.sec = sec

  def addMonth(self, a):
    return MyDate(self.year, self.month + a, self.date, self.hour, self.min, self.sec)

  def show(self):
    print(f'{self.year}-{self.month}-{self.date}{self.hour}:{self.min}:{self.sec}')
```

```python
startDate = MyDate(2023, 2, 9, 19, 55, 5)
finshDate = MyDate(2023, 2, 10)

startDate.show()
res = startDate.addMonth(3)
res.show()
```

#**4. object 클래스**

* 파이썬의 모든 클래스는 object 클래스를 상속 받는다.

```python
# 2차원 좌표 클래스

class Coord:
  def __init__(self, x, y):
    self.x = x
    self.y = y

def __add__(self, a):
  res1 = self.x + a.x
  res2 = self.y + a.y
  return Coord(res1, res2)
  
# x 좌표와 y좌료를 변경하는 flip

def flip(self):
  self.x, self.y = self.y, self.x


def show(self):
  print(f'2차원 평면 좌표[{self.x}, {self.y}]')
```

```python
a = Coord(3, 1)
b = Coord(2, 5)
res = a + b
res.show()
```

```python
a = Coord(3, 1)
a.show()
a.flip()
a.show()
```

#**5. 클래스 변수와 클래스 메소드, 스태틱 메소드

* 인스턴스변수 : 인스턴스마다 저장된 값이 달라지는 변수  
* 클래스 변수 : 인스턴스 마다 저장된 값이 달라지지 않는다(같은 클래스의 인스턴스라면 모두 공유하는 변수)  

```python
class Student:
    # def __init__(self, name, age):
    # 인스턴스변수 name 생성
    school = 'IT 대학교'

    def __init__(self, name, age):

      self.name = name

    # 인스턴스변수 age 생성

      self.age = age

    @classmethod      # 어노테이션 주석, : 컴퓨터 한테하는 설명
    def intro(cls):
      print(cls)
      print('클래스 메소드!')
      print(f'전체 학생수 : {cls.cnt}')

    # 일반 메소드
    def show(self):
      print(self) 
```

```python
kim = Student('박준', 30)
hong = Student('준박', 31)

Student.school = '고등학교'

print(Student.school)
print(kim.school)
print(hong.school)

print(Student.cnt)

Student.intro()   # 클래스 메소드와 클래스 변수는 클래스 이름으로 바로 사용하는 것이 바람직
kim.intro()
kim.show()
```

```python
# static 메소드 메모리에 한번 올라가는 메소드
# static 메소드는 매개 변수로 전달 되는 값이 없다.
# 유틸리티함수(기능 함수)

class math:
  @staticmethod
  def add():
    print('math 클래스 안의 add() 실행됨')

class econ:
  @staticmethod
  def add():
    print('econ 클래스 안의 add() 실행됨')
```

```python
math.add()
econ.add()
```
