# **0. 연산자**

1. 피연산자의 타입
2. 연산의 동작
3. 연산 결과 값의 타입



연산자의 해석은 연산 결과 값으로 **치환** 하여 해석한다.


# **1. 산술 연산자**

1. \+ : 덧셈 연산자
> str + str --> str, 연산의 결과는 두 문자열이 연결된 하나의 str 값이다.  
> 피연산자 자리에 어느 한쪽에라도 float 타입이 있다면 결과는 float 타입이다.  
> int + int --> int

2. \- : 뺄셈 연산자
3. \* : 곱셈 연산자
4. / : 나눗셈 연산자
5. // : 몫 연산자
6. % : 나머지 연산자

## **1-1. 덧셈 연산자**

```python
10 + 5 # 15, int + int --> int

10 + 3.14  # 13.14, int + float --> float

# int + str, str + int --> 오류
# 10 + '안녕' # TypeError: unsupported operand type(s) for +: 'int' and 'str'


3.5 + 1.5 # 5.0, float + float --> float

# float + str, str + float --> 오류
# 3.5 + '10' #TypeError: unsupported operand type(s) for +: 'float' and 'str'

'안녕' + '반가워' # '안녕 반가워', str + str --> str 연결된다. (연산 결과는 문자열)

```

type(값) : 괄호 안에 있는 값의 타입을 결과로 가져오는 함수

함수 사용의 결과는 타입이 된다.

```python
print (type(10))
print (type('안녕'))

```

## **1-2. 뺄셈 연산자**

* str은 뺄셈 불가능
* 피연산자 어느 한 쪽이라도 float 타입이 있다면 결과는 float 타입
* int - int --> int

```python
type(10 - 5) # 5, int - int --> int
type(10 - 0.0) # 10.0, int - float, float - int --> float

# str은 뺄셈 연산 불가능
# type(10 - '10')

type(1.2 - 0.2) # 1.0, float - float --> float

```

## **1-3. 곱셈 연산자**


* str 타입은 int랑 만 가능
* 피연산자 자리에 어느 한 쪽에라도 float 타입이 있다면 결과는 float 타입이 나온다.  
* int*int --> int

```python
type('안녕' * 3) # '안녕안녕안녕' --> str * int int * str --> str, 반복

# '안녕' * 3.5 # 오류

type(3*3) # int * int --> int
type(3.5 * 0) # int * float, float * int --> float
type(1.2 * 1.2) # float * float --> float
```

## **1-4. 나눗셈 연산자**

* str은 사용 불가능
* 연산 결과는 무조건 float 타입
* 0으로 나누면 ZeroDivisionError 발생



```python
10 / 3 # int / int
```

## **1-5. 기타 연산자**

* // : 몫 연산자, 몫을 계산한다.
* % : 나머지 연산자, 나머지를 계산한다.
* ** : 파워 연산자, 거듭 제곱을 계산한다.

```python
10 // 3 # 결과는 3, 몫을 계산

10 % 3 # 결과는 1, %는 나머지(나누기에서 남은 값을 계산함.)

10 ** 3
```

#**2. 형변환 함수**

* int() : 괄호 안에 있는 값을 int 타입을 형변환
* float() : 괄호 안에 있는 값을 float 타입으로 형변환\
* str() : 괄호 안에 있는 값을 str 타입으로 형변환

함수 사용 결과를 형변환된 값으로 **치환** 하여 생각한다.

```python
# int()
type (int('25')) # 괄호안에 있는 값이 int 타입으로 변환
int(3.14) # 괄호안에 있는 값이 int 타입으로 변환

# float()
float('5.12') # str --> float 타입으로
float(5) # int --> float 타입으로

str(25)
str(3.14)
```

## **3-1. 사용자에게 정수를 입력 받고, 해당 정수를 5로 나우었을 때 몫과 나머지를 출력하시오**


입력 예시  
  정수 입력 >> 27  
      
    
출력 예시  
  27을 5로 나누었을 때 몫은 5이며, 나머지는 2입니다.

```python
num = int(input('정수 입력 >> '))

print(num, '을 5로 나누었을 때 몫은', num // 5, '이며, 나머지는', num % 5, '입니다')

print(num, '을 5로 나누었을 때 몫은', num // 5, '이며, 나머지는', num % 5, '입니다', sep='')
```

##**3-2. 사용자에게 두 자릿수 정수를 입력 받고, 해당 정수의 10의 자리와 일의 자리를 출력하시오**

입력 예시  
두자릿수 정수 입력 >> 25

출력 예시  
십의 자리 : 2  
일의 자리 : 5

```python
num = int(input('두자릿수 정수 입력 >>'))

print('십의 자리 : ', num // 10)
print('일의 자리 : ', num % 10)

```

## **3-3. 노래 한 곡당 300원인 코인 노래방이 있다. 사용자에게 금액을 입력 받고, 부를 수 있는 노래 곡 수와 잔돈을 출력하시오**

입력 예시  
금액 입력 >> 1600  

출력 예시
  부를 수 있는 노래 : 5곡  
  잔돈 : 100원  


```python
money = int(input('금액 입력 >> '))


price = 300


songCnt = money // price # songCnt에는 노래 곡수가 저장됨
change = money % price # change에는 잔돈이 저장됨

print('부를 수 있는 노래 : ', songCnt, '곡' , sep = '')
print('잔돈 : ' , change, '원', sep = '')

```

## **4-1. bool 타입 형변환**

* bool() 함수를 사용한다.

```python
# int 타입의 값을 bool 타입으로 바꾸면 0은 False 0이 아닌 수는 True로 바뀐다.

bool(10) # True
bool(-5) # True
bool(0) # False

# float 타입의 값을 bool 타입으로 바꾸면 0.0은 False 그외의 수는 True로 바뀐다.

bool(3.14) # True
bool(-5.1) # True
bool(0.0) # False

# str 타입의 값을 bool 타입으로 바꾸면 비어있는 문자열 '' 만 False, 그 외의 문자열은 모두 True로 바뀐다.

bool('True') # True
bool('False') # True
bool('') # False

# None

type(None) # NoneType : 값이 없음을 나타내는 값

bool(None) # False
```

## **4-2. 관계(비교) 연산자**
    a > b : a가 b 보다 크니? 맞으면 True 아니면 False
    a < b : a가 b 보다 작니? 맞으면 True 아니면 False
    a >= b : a가 b 보다 크거나 같니? 맞으면 True 아니면 False
    a <= b : a가 b 보다 작거나 같니? 맞으면 True 아니면 False
    a == b : a가 b 랑 서로 같니? 맞으면 True 아니면 False
    a != b : a가 b 보다 서로 다르니? 맞으면 True 아니면 False

관계 연산의 결과는 bool 타입 값이다.

```python
5 > 12.2 # False
3.0 == 3 # True
'3' == 3 # False
'3' == 51

'apple' < 'banana' # True
'apple' >= 'angel'# True
'안녕' == '안녕' # True

True > False # True
True == 1 # True는 1 False는 0으로 생각하여 비교
```

# **5. 논리연산자**
    a and b : a가 True 그리고 b 도 True라면 결과는 True 아니면 False
    a or b : a 혹은 b 둘 중 하나라도 True라면 결과는 True 아니면 False
    not a : a가 True라면 결과는 False a가 False라면 결과는 True

```python
True and True # True
True and False # False
False and True # False
False and False # False
```

```python
True and True # True
True and False # True
False and True # True
False and False # False
```

```python
not True # False
not False # True
```

```python
True or True

not (10 % 3 == 0) and 3 != 50 - 20 # True
```

## **5-1. 논리 연산자 심화**

논리 연산의 결과는 맥락 평가에 따라서 나온다.

```python
10 or 0 # 피연산자 자리에 bool이 아닌 다른 타입 값이 쓰였다면 그 값을 bool()타입으로 변환하여 해석합니다.

0 or 20

'안녕' and ''
'' and 15
```
