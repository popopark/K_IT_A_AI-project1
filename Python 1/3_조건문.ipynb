# **1. 조건문**

조건에 따라서 다른 코드를 실행시키고자 할 때 (분기 처리 할 때) 사용한다.

##**if문**

  if문 바깥쪽 문장  
  if 조건식 :   
      if문 안쪽 문장  
      if문 안쪽 문장  
  if문 바깥쪽 문장

  if문을 만나면 1. 조건식을 확인한다.

  1-1. 만약 조건식이 True(bool 타입으로 변환 했을때 True) 라면
  if문 안쪽으로 들어와서 안쪽 문장을 차례로 실행한다.

  1-2. 만약 조건식이 False(bool 타입으로 볍ㄴ환 했을 때 False)
  라면 if문 안쪽 문장을 실행하지 않는다.

  




```python
print('프로그램을 시작 했어요')       # 첫 번째 줄

if False:                              # 두 번째 줄 / if문이 들여쓰기를 통해 영역을 알기 때문에 중요하다.
  print('if문 안쪽 문장입니다1')
  print('if문 안쪽 문장입니다1')

print('프로그램을 시작 했어요')       # 세 번째 줄

```

```python
# 사용자에게 정수를 입력 받아서 짝수라면 짝수입니다.
# 홀수라면 홀수 입니다 출력하기

print('프로그램 시작')

num = int(input('정수 입력 >>'))

if num % 2 == 0:
  print('짝수입니다.')
  
if num % 2 != 0:
  print('홀수입니다.')

print('프로그램 끝')
```

##**if~else**

    if 조건식:
      조건식이 참일때 실행 할 문장
      ...
    else :
      조건식이 거짓일 때 실행 할 문장
      ...


```python
# 특정 조건이 거짓 일 때 무언가를 실행하고 싶다면 else를 사용 할 수 있다.
# if ~ else는 하나의 조건문, else로 시작 할 수는 없다.


num = int(input('정수 >>'))   # 컴퓨터 입장도 생각해라


if num % 2 == 0:
  print('짝수입니다')

else :                        # else는 거짓 일 때 뭔가를 실행 시키고 싶을 때 씀 (조건식이 거짓 일 때)
  print('홀수입니다')

print('프로그램 끝')
```

##**if문과 삼항연산자**

삼항 연산자는 코드를 간결히 작성 할 수 있지만 가독성이 떨어진다.

if문은 가독성이 좋다.
모든 삼항 연산자는 if문으로 바꿀 수 있다.

하지만 모둔 if문을 삼항 연산자로 바꿀 수는 없다.

```python
# 두 수를 입력 받아서 둘 중 더 큰 수 혹은 서로 같다면 같다고 출력하기

num1 = int(input('정수 1 >>'))
num2 = int(input('정수 2 >>'))

num1 if num1 > num2 else num2 if num1 < num2 else '같다'
```

```python
# 두 수를 입력 받아서 둘 중 더 큰 수 혹은 서로 같다면 같다고 출력하기

num1, num2 = list(map(int, input('두 수 띄어쓰기로 입력 >>').split()))

print(num1 if num1 > num2 else num2 if num1 < num2 else '같다')

# if문으로 변경

if num1 > num2 :
  print(num1)
else :
  #--------------------------
  if num2 > num1:
    print(num2)
  else :
    print('같다')
  #--------------------------
#--------------------------
```

## **if ~ elif ~ else**
    if 조건식 1:
      조건식 1이 참 알때 실행되는 영역
    else 조건식 2:
      조건식 1은 거짓, 2는 참 일때 실행되는 영역
    elif 조건식 3:
      조건식1,2는 거짓, 3은 참 일때 실행되는 영역
    ...:
      ...
    else:
      위에있는 조건식이 모두 거짓일때 실행되는 영역

```python
num1, num2 = list(map(int, input().split()))

if num1 > num2 :
  print(num1)

elif num2 > num1 :
  print(num2)

else :
  print('같다')

print('반가워')

print('프로그램 종료')
```

## **예제1**

시험 점수를 입력 받아 올바른 학점을 출력하시오

* 90점 이상 : A
* 80점 이상 90미만 : B
* 70점 이상 80미만 : C
* 70점 미만 : F

당신의 학점은 0입니다.

```python
score = int(input('시험 점수 >>'))

if score >= 90:
  print('당신의 학점은 A입니다.')

elif 80 <= score < 90:
  print('당신의 학점은 B입니다.')

elif 70 <= score < 80:
  print('당신의 학점은 C입니다.')

else:
  print('당신의 학점은 F입니다.')
```

```python
score = int(input('시험 점수 >>'))

grade = 'F'

if score >= 90:
  grade = 'A'

elif score >= 80:
  grade = 'B'

elif score >= 70:
  grade = 'C'

print(f'당신의 학점은 {grade}입니다.')
```

## **예제 2**

사용자에게 년도를 입력 받아서 윤년인지 아닌지 출력하기

윤년의 조건은 다음과 같다.

* 4의 배수 0 100의 배수 0 400의 배수 O --> 윤년
(2000, 2400)

* 4의 배수 0 100의 배수 0 400의 배수 X --> 평년
(1900, 2100)

* 4의 배수 0 100의 배수 x 400의 배수 X --> 윤년
(2020, 2024)

* 4의 배수 X 100의 배수 X 400의 배수 X --> 평년
(1900, 2100)

```python
# 4의 배수 0 : 숫자 % 4 == 0
# 100의 배수 0 : 숫자 % 100 == 0
# 400의 배수 0 : 숫자 % 400 == 0

year = int(input('년도 입력 >>'))

if year % 4 == 0 and year % 100 == 0 and year % 400 == 0:
  print('윤년')

if year % 4 == 0 and year % 100 == 0 and year % 400 != 0:
  print('평년')

if year % 4 == 0 and year % 100 != 0 and year % 400 != 0:
  print('윤년')

if year % 4 != 0 and year % 100 != 0 and year % 400 != 0:
  print('평년')

```

# **1. for문**

  for 변수명 in 반복 가능 객체:  
    반복 할 코드  
    반복 할 코드  
    ...

```python
print('프로그램 시작')

for i in [10, 20, 30]:
  print('박준')
  print('for 문 안쪽!')

print('for 문 아래쪽!')
```

```python
# range(a) : range() 함수 사용 결과는 0부터 a-1까지의 숫자들이
#           요소로 들어있는 반복 가능 객체(range 타입)


list (range(10))
```

```python
# range(a) : range() 함수 사용 결과는 0부터 a-1까지의 숫자들이
#           요소로 들어있는 반복 가능 객체(range 타입)

# range(start, end) : range() 함수 사용 결과는 start부터 end-1 까지의 숫자들이
#           요소로 들어있는 반복 가능 객체(range 타입)

# range(start, end, step) : range()함수 사용 결과는 start부터 end-1 까지 step 간격의 숫자들이
#           요소로 들어있는 반복 가능 객체

print(list(range(1, 10, 3)))

list (range(10))

for i in range(5):
  print('박준')
```

##**1-1. for 문의 변수 자리**

  for 변수명 in 반복 가능 객체:
    반복 할 코드

* 반복의 횟수 : 반복 가능 객체의 요소 갯수
* 변수에 대입되는 값 : 반복 가능 객체의 요소들이 차례로 대입된다.

```python
# range(5) --> 0, 1, 2, 3, 4가 요소로 들어있는 반복 가능 객체


for i in range(5):

  print(i)

print('-----------')

for i in [10, 20, 30, 40]:
  print(i)

print('-'*20)

for i in (5, 6, 7, 8):
  print(i)

print('-'*20)

for i in {'박준', '김준', '최준', '성준'}:
  print(i)

print('-'*20)       # 딕셔너리 출력

scores = {'국어':50, '수학':60, '영어':70}

for i in scores:
  print(scores[i])
```

#**2. continue와 break**

* continue : 반복문 안에서 사용되며,
cotinue가 실행되면 즉시 다음 반복을 실행한다.

* break : 반복문 안에서 사용되며, break가
실행되면 즉시 반복을 종료한다.

```python
for i in range(5):

  print(f'현재 i값 : {i}')
  if i % 2 == 0:
    continue
  print('for문 안쪽 문장')

print('for문 바깥 문장!')

```
