```python
def stInfo(a):
  print('----학생 정보----')
  print(f'이름 : {a.name}')
  print(f'나이 : {a.age}')  
```

```python
# name = str()

# num = int()
# print(num)

# s1 객체 만들기
s1 = Student()

type(s1)
s1.name = '박준'
s1.age = 10     # . (점)은 하위 연산자임

s2 = Student()
s2.name = '박준'
s2.age = 30

print(s1)
s1.show()
s2.show()       # 0x7fa8f63c16a0 파란색 객체에 대한 정보임

# list 함수는 묶어서 조금 더 쉽게 정리해주는 함수다.

stInfo(s1)

# print('---학생 정보---')
# print(f'이름 : {s1.name}')
# print(f'나이 : {s1.age}')

stInfo(s2)

# print('---학생 정보---')
# print(f'이름 : {s2.name}')
# print(f'나이 : {s2.age}')
```
