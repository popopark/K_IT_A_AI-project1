# **5. 배열**

* 데이터를 나열하고 각 데이터를 인덱스에 대응하도록 구성한 자료 구조  
* 파이썬 리스트 타입은 배열 기능을 제공함

### **5-1. 배열이 필요한 이유**

* 같은 종류의 데이터를 효율적으로 관리하기 위해 사용
* 같은 종류의 데이터를 0번 부터 순차적으로 저장
* 빠른 접근이 가능(인덱스 번호로 접근시)  

### **5-2. 배열의 단점**

* 미리 최대 길이를 설정해야 함
* 데이터의 추가/삭제가 어려움

```python
# 1차원 배열 : 리스트로 구현

arr = [1, 2, 3, 4, 5]
print(arr)
```

```python
# 2차원 배열 : 리스트로 구현

arr = [[1,2,3,], [4,5,6], [7,8,9]]
print(arr)
```

```python
# 인덱싱

print(arr[0])
print(arr[0][0])
print(arr[1][2])
```

```python
# arr에서 9 8 7 순서로 값을 출력

# arr = [9, 8, 7]
# print(arr)

print(arr[2][2], arr[2][1], arr[2][0])
print(arr[2][::-1])

# Extened Slices
# arr[::-1]
```

### **5-3. Extened Slices**

* arr[A:B:C] : index A부터 index
B까지 C의 간격으로 배열을 생성
* A가 None인 경우 : 처음부터
& B가 None인 경우 : C가 양수일 경우 마지막 index까지, C가 음수라면 첫 index까지  

```python
arr = range(10)

print(arr)
print(arr[1:9:2])
print(arr[:9:2])
print(arr[::-1])

# list(arr[1:9:2])
```

# **6. 큐(Queue)**

* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조
* 줄을 서는 행위와 유사
* FIFO(First-ln, First-Out)

### **6-1. 큐의 사용**

* 멀티 테스킹을 위한 프로세스 스케쥴링 방식을 구현(운영체제)  
* 푸시 메세지

###**6-2. 큐의 용어**

* Enqueue : 큐에 데이터를 넣는 기능
* Dequeue : 큐에 데이터를 꺼내는 기능 
* [VisualGO](https://visualgo.net/en/list)

### **6-3. queue 라이브러리를 활용**

* Queue(): 가장 일반적인 큐 자료 구조를 생성
* LifoQueue() : 나중에 입력된 데이터가 먼저 출력되는 구조
* PriorityQueue() : 데이터마다 우선 순위를 넣어서 우선 순위가 높은 순으로 데이터를 출력 

```python
import queue

data_queue = queue.Queue()
data_queue.put('apple')  # Enqueue
print(str(data_queue))
print(data_queue.qsize())
data_queue.put('banana') # Enqueue
print(data_queue.qsize())
print(data_queue.get())  # Dequeue
print(data_queue.qsize())
print(data_queue.get())  # Dequeue
print(data_queue.qsize())
```

```python
li = [1, 2, 3, 4]
print(li)
print(type(li))
```

```python
data_queue.put('apple')
data_queue.put('김사과')
data_queue.put(183.5)
data_queue.put(True)
print(data_queue.qsize())
```

```python
for _ in range(data_queue.qsize()):   # 안쓰는 변수는 _ (언더바)를 써주는게 좋음
  item = data_queue.get()
  print(item)
```

```python
data_queue = queue.LifoQueue()
data_queue.put('apple')
data_queue.put('박준')
data_queue.put('183.9')
data_queue.put('True')
print(data_queue.qsize())
print(data_queue.get())
```

```python
# 숫자가 작을수록 우선 순위가 높음

data_queue = queue.PriorityQueue()
data_queue.put((10, 'apple'))
data_queue.put((9, 'aple'))
data_queue.put((7, 'ape'))
data_queue.put((5, 'le'))
data_queue.put((3, 'e'))
print(data_queue.qsize())
print(data_queue.get())
print(data_queue.get())
```

### 문제

리스트 변수로 Queue를 다루는 enqueue, dequeue 기능을 직접 구현해보자

```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]     # del은 지우는 것임
  return data
```
